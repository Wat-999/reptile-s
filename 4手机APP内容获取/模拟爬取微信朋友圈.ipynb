```python
import time

data_all = ''
for i in range(20):  # 这里设置向下翻的次数，可以自定义调节
    data_old = browser.page_source  # 滑动前旧的网页源代码
    data_all = data_all + data_old
    
    browser.swipe(50, 1000, 50, 200)  # 往下翻
    time.sleep(2)  # 翻页完歇2秒
    
    data_new = browser.page_source  # 滑动后新的网页源代码
    
    if data_new == data_old:  # 如果网页源代码没有变化，说明滑到头了
        break  # break语句退出for循环
    else:
        pass  # pass表示什么也不干，继续循环
```

```python
data_all = ''
for i in range(20):
    data_old = browser.page_source
    data_all = data_all + data_old
    a = browser.find_elements_by_id('com.tencent.mm:id/gbx')
    for i in a:
        print(i.text)
    
    browser.swipe(50, 1000, 50, 200)  # 往下翻
    time.sleep(2)  # 翻页完歇2秒
    
    data_new = browser.page_source
    
    if data_new == data_old:
        break
    else:
        pass
```

```python
data_all = ''
text_all = []  # 1.构造一个空列表，存储文本信息
for i in range(20):
    data_old = browser.page_source
    data_all = data_all + data_old
    a = browser.find_elements_by_id('com.tencent.mm:id/gbx')
    for i in a:
        text_all.append(i.text)  # 2.通过append()函数
    
    browser.swipe(50, 1000, 50, 200)
    time.sleep(2)
    
    data_new = browser.page_source
    
    if data_new == data_old:
        break
    else:
        pass

text_all = set(text_all)  # 3.通过set()函数进行去重
for i in text_all:  # 4.打印去重后的内容
    print(i)
```
